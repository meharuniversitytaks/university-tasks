```python
import pandas as pd
import numpy as np
df=pd.read_csv("melb_data.csv")
```

```python
new=df.isna().sum()
```

```python
percentage=new/len(df) * 100
```

```python
percentage
```

```python
 b=percentage[percentage>20].index
```

```python
df=df.drop(b,axis=1)
```

```python
df.head(3)
```

```python
#----2
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
categorial_columns=df.select_dtypes(include=["object"]).columns
```

```python
for col in categorial_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```
